# Creating first Dataframe to be used

## Importing necessary libraries

```python
import pandas as pd
```

## Reading dataframe to be used and modified

```python
original_df = pd.read_pickle('FallAllD.pkl')     # Reading original Dataset
original_df.head(10)
```

## Adding a new column to the original dataframe

This new column is for classifying whether the row represents a fall or not. If Activity ID is greater than 100, it represents a fall.

```python
dataframe1 = original_df.copy()         # Creating a copy of original datset into the variable
dataframe1['Fall'] = 0                  # Creating a new column with Fall, which become our output
for x in dataframe1.index:
    if (dataframe1.iloc[x, 2] > 100):           # Checking whether the row represents a Fall or ADL, second index refers to ActivityID
        dataframe1.iloc[x, -1] = 1              # If activity ID is above 100, the row represents a Fall
    else:
        dataframe1.iloc[x, -1] = 0              # Else, row represents a ADL

print('------------------HEAD------------------------------')
print(dataframe1.head())
print('------------------TAIL------------------------------')
print(dataframe1.tail())
```

## Converting data in columns and creating new pickle

Multiplying data in Acc, Gyr and Mag columns by the sensitivity of the sensor, to get the real data

Accelerometer sensitivity 0.244 mg/LSB
Angular rate sensitivity 70 mdps/LSB
Magnetic sensitivity 0.14 mgauss/LSB

```python
dataframe1['Acc'] = dataframe1['Acc'] * 0.000244        # Since values are not stored with original values, this becomes useful
dataframe1['Gyr'] = dataframe1['Gyr'] * 0.07            # to modify data to become the same as another sensor used by another project
dataframe1['Mag'] = dataframe1['Mag'] * 0.00014         # so to get real data we will multiply each value by the respective sensor sensitivy

dataframe1.to_pickle('Dataframe1.pkl')
```

# Creating second Dataframe to be used

```python
dataframe2 = dataframe1.copy()          # Creating a copy of dataframe1 into variable dataframe2
dataframe2['ADL'] = 2                   # Creating a new column for identifying ADL

for x in dataframe2.index:
    if(dataframe2.iloc[x,-2] == 1):     # Checking if Fall column is 1 or 0
        dataframe2.iloc[x,-1] = 0       # If Fall = 1, then ADL needs to be the opposite
    else:
        dataframe2.iloc[x,-1] = 1       # If Fall = 0, then ADL is equal to 1

dataframe2.to_pickle('Dataframe2.pkl')
```

## Removing certain activities to have a better balance

The balanced dataframe contains a Fall and ADL column, with just 693 ADLs and 466 Falls.

```python
balancedDataframe = dataframe2
IDtoRemove = [44,43,19,21,29,31,32,34,15,16,17,22,24,35,36,26,42,41]

for x in IDtoRemove:
    balancedDataframe.drop(balancedDataframe[balancedDataframe.ActivityID == x].index, inplace=True)

balancedDataframe.reset_index(drop=True, inplace=True)  # Resetting index after removing "unwanted" ADLs    
balancedDataframe.loc[balancedDataframe['ADL'] == 1].count()        # Getting the amount of ADL in the dataframe
balancedDataframe.info()
```

```python
balancedDataframe.to_pickle('Balanced Dataframe.pkl')
```

# Creating third Dataframe to be used

This dataframe will have just 1 column, but we will be using average, std deviation and range.

```python
dataframe3 = pd.read_pickle('Dataframe1.pkl')

dataframe3['Mean Acc X'] = -1
dataframe3['Mean Acc Y'] = -1
dataframe3['Mean Acc Z'] = -1
dataframe3['Std Acc X'] = -1
dataframe3['Std Acc Y'] = -1
dataframe3['Std Acc Z'] = -1
dataframe3['Rng Acc X'] = -1
dataframe3['Rng Acc Y'] = -1
dataframe3['Rng Acc Z'] = -1

dataframe3['Mean Gyr X'] = -1
dataframe3['Mean Gyr Y'] = -1
dataframe3['Mean Gyr Z'] = -1
dataframe3['Std Gyr X'] = -1
dataframe3['Std Gyr Y'] = -1
dataframe3['Std Gyr Z'] = -1
dataframe3['Rng Gyr X'] = -1
dataframe3['Rng Gyr Y'] = -1
dataframe3['Rng Gyr Z'] = -1

dataframe3['Mean Mag X'] = -1
dataframe3['Mean Mag Y'] = -1
dataframe3['Mean Mag Z'] = -1
dataframe3['Std Mag X'] = -1
dataframe3['Std Mag Y'] = -1
dataframe3['Std Mag Z'] = -1
dataframe3['Rng Mag X'] = -1
dataframe3['Rng Mag Y'] = -1
dataframe3['Rng Mag Z'] = -1
''''''
for x in dataframe3.index:
    
    AccX = pd.DataFrame(dataframe3.iloc[x].Acc.tolist())[0]
    AccY = pd.DataFrame(dataframe3.iloc[x].Acc.tolist())[1]
    AccZ = pd.DataFrame(dataframe3.iloc[x].Acc.tolist())[2]
    
    GyrX = pd.DataFrame(dataframe3.iloc[x].Gyr.tolist())[0]    
    GyrY = pd.DataFrame(dataframe3.iloc[x].Gyr.tolist())[1]
    GyrZ = pd.DataFrame(dataframe3.iloc[x].Gyr.tolist())[2]

    MagX = pd.DataFrame(dataframe3.iloc[x].Mag.tolist())[0]    
    MagY = pd.DataFrame(dataframe3.iloc[x].Mag.tolist())[1]
    MagZ = pd.DataFrame(dataframe3.iloc[x].Mag.tolist())[2]

    dataframe3.iloc[x, 8] = AccX.mean()
    dataframe3.iloc[x, 9] = AccY.mean()
    dataframe3.iloc[x, 10] = AccZ.mean()
    dataframe3.iloc[x, 11] = AccX.std()
    dataframe3.iloc[x, 12] = AccY.std()
    dataframe3.iloc[x, 13] = AccZ.std()
    dataframe3.iloc[x, 14] = AccX.max() - AccX.min()
    dataframe3.iloc[x, 15] = AccY.max() - AccY.min()
    dataframe3.iloc[x, 16] = AccZ.max() - AccZ.min()

    dataframe3.iloc[x, 17] = GyrX.mean()
    dataframe3.iloc[x, 18] = GyrY.mean()
    dataframe3.iloc[x, 19] = GyrZ.mean()
    dataframe3.iloc[x, 20] = GyrX.std()
    dataframe3.iloc[x, 21] = GyrY.std()
    dataframe3.iloc[x, 22] = GyrZ.std()
    dataframe3.iloc[x, 23] = GyrX.max() - GyrX.min()
    dataframe3.iloc[x, 24] = GyrY.max() - GyrY.min()
    dataframe3.iloc[x, 25] = GyrZ.max() - GyrZ.min()

    dataframe3.iloc[x, 26] = MagX.mean()
    dataframe3.iloc[x, 27] = MagY.mean()
    dataframe3.iloc[x, 28] = MagZ.mean()
    dataframe3.iloc[x, 29] = MagX.std()
    dataframe3.iloc[x, 30] = MagY.std()
    dataframe3.iloc[x, 31] = MagZ.std()
    dataframe3.iloc[x, 32] = MagX.max() - MagX.min()
    dataframe3.iloc[x, 33] = MagY.max() - MagY.min()
    dataframe3.iloc[x, 34] = MagZ.max() - MagZ.min()
    
dataframe3.to_pickle('Dataframe with Statistical Characteristics.pkl')
```

```python
col = dataframe3.pop('Fall')

dataframe3.insert(34, col.name, col)

dataframe3.tail()

dataframe3.to_pickle('Dataframe3.pkl')
```
