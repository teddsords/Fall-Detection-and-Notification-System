## Adding a new column to the original dataframe

This new column is for classifying whether the row represents a fall or not

```python
dataframe1 = originalDF
#dataframe1['Fall'] = 0
#dataframe1.head(10)

for x in dataframe1.index:
    if (dataframe1['ActivityID'][x] > 100):
        dataframe1['Fall'][x] = 1
    else:
        dataframe1['Fall'][x] = 0

print('------------------HEAD------------------------------')
print(dataframe1.head())
print('------------------TAIL------------------------------')
print(dataframe1.tail())
```
