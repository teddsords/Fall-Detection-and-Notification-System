## Sorting/Describing data

```python
df.describe()
```

```python
print(df.sort_values('Name'))       # use sort_values funciton to sort them by the column chosen
```

```python
print(df.sort_values('Name', ascending=False))      # Sort values in descending order
```

```python
print(df.sort_values(['Type 1', 'HP']))             # Sorting by two or more columns at the same time
```

```python
print(df.sort_values(['Type 1', 'HP'], ascending= False))       # Both columns will be in descending order
```

```python
print(df.sort_values(['Type 1', 'HP'], ascending=[True, False]))        # But we can specify which column will be ascending and which will be descinding
```

## Making changes to the data

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']     # Creating a new column by adding every category
df = df.drop(columns = ['Total'])                                                                       # Dropping a column
df.head(5)
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis = 1)           # Another way to add a new column, in iloc we say we need every row and specify which columns will be added, axis = 1 for horizontal sum
print(df.head(5))
```

```python
cols = list(df.columns.values)
df = df[cols[0:4] + [cols [-1]] + cols[4:12]]       # Rearrging columns, so that total colummn is in the middle. Be careful when hard coding with numbers
print(df.head())
```
