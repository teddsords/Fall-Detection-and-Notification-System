## Making changes to the data

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']     # Creating a new column by adding every category
df = df.drop(columns = ['Total'])                                                                       # Dropping a column
df.head(5)
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis = 1)           # Another way to add a new column, in iloc we say we need every row and specify which columns will be added, axis = 1 for horizontal sum
print(df.head(5))
df.to_csv('modified.csv')
```

```python
cols = list(df.columns.values)
df = df[cols[0:4] + [cols [-1]] + cols[4:12]]       # Rearrging columns, so that total colummn is in the middle. Be careful when hard coding with numbers
print(df.head())
```

## Aggregate Satistics (Gropby)

```python
df.groupby(['Type 1']).mean()           # Getting stats of each sorted by type 1 pokemon
```

```python
df.groupby(['Type 1']).mean().sort_values('Defense' , ascending=False)      # Same as before, but sorting row by its defense value in descending order
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
df['count'] = 1

df.groupby(['Type 1']).count()['count']             # We are grouping by Type 1 and counting how many there are for each one of them, but it will only count the 'count' column
```
